# Model Validation Techniques
##Two Techniques: 1. Collinearity Check
## 2. Residual Analysis

```python
# 1. Collinearity problem Check
# Calculating VIF =1/(1-Rsquare) for all independent variables

rsq_RD= smf.ols('RD~Admin+Market', data= startups).fit().rsquared
vif_RD= 1/(1-rsq_RD)

rsq_AD= smf.ols('Admin~RD+Market', data= startups).fit().rsquared
vif_AD= 1/(1-rsq_AD)

rsq_MA= smf.ols('Market~RD+Admin', data= startups).fit().rsquared
vif_MA= 1/(1-rsq_MA)

# Putting the values in Dataframe format
d1={'Variables':['RD','Admin','Market'],'vif':[vif_RD,vif_AD,vif_MA]}
vif_df=pd.DataFrame(d1) 
vif_df
```

```python
# No variable has VIF>5, No Collinearity, so consider all variables in Regression equation
```

```python
# 2) Residual Analysis
# Test for Normality of Residuals (Q-Q Plot) using residual model(model.resid)
sm.qqplot(model.resid, line='q') # 'q' -A line is fit through the quartiles # line ='45'- to draw the 45-degree diagonal line
plt.title('Normal Q-Q plot of residuals')
plt.show()
```

#Model Deletion Diagnostics (checking Outliers or Influencers)
##Techniques:
##1. Cook's Distance
##2. Leverage value

```python
# 1. Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
(c,_)= model.get_influence().cooks_distance
c
```

```python
# Plot the influencers using the stem plot
fig= plt.figure(figsize= (20,7))
plt.stem(np.arange(len(startups)), np.round(c,3))
plt.xlabel('Row Index')
plt.ylabel("Cook's Distance")
plt.show();
```

```python
# Index and value of influencer where C>0.5
np.argmax(c), np.max(c)
```

```python
# 2. Leverage Value using High Influence Point's : Points beyond Leverage_cutoff value are influencers
fig, ax= plt.subplots(figsize = (15,10))
fig= influence_plot(model, ax=ax)
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k= startups.shape[1]
n= startups.shape[0]
leverage_cutoff= (3*(k+1))/n
leverage_cutoff
```

# Improving the Model

```python
# Creating a copy of data so that original dataset is not affected
startups_new= startups.copy()
startups_new
```

```python
# Discard the data points which are influencers and reassign the row number (reset_index(drop=True))
startups1= startups_new.drop(startups_new.index[[49]], axis= 0).reset_index(drop=True)
startups1
```

# Model Deletion Diagnostics and Final Model

```python
while np.max(c)>0.5 :
  model= smf.ols('Profit~RD+Admin+Market', data= startups1).fit()
  (c,_)= model.get_influence().cooks_distance
  c
  np.argmax(c), np.max(c)
  startups1= startups1.drop(startups1.index[[np.argmax(c)]], axis= 0).reset_index(drop=True)
  startups1
else:
  final_model= smf.ols('Profit~RD+Admin+Market', data= startups1).fit()
  final_model.rsquared, final_model.aic
  print("Thus model accuracy is improved to", final_model.rsquared)
```
