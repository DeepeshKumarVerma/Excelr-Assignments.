```python
books_new= books1.pivot_table(index='UserId',columns= 'BookTitle', values= 'Rating')
books_new.head()
```

```python
# Imputing Those NaN with 0 values
books_new.fillna(0, inplace= True)
```

```python
books_new.index= books1['UserId'].unique()
books_new
```

# Calculating Cosine Similarity between Users

```python
user= 1-pairwise_distances(books_new.values, metric='cosine')
user
```

```python
np.fill_diagonal(user,0)
```

```python
# Store the results in a dataframe
user1= pd.DataFrame(user)
user1
```

```python
# Set the index and columns names to user ids
user1.index= books1.UserId.unique()
user1.columns= books1.UserId.unique()
```

### Considering the most similar customer

```python
def recommend(custID):
  simID = user.loc[custID].idxmax()
  simID_books= books1[books1['UserId']== simID].BookTitle 
  custID_books= books1[books1['UserId']== custID].BookTitle
  return set(simID_books)- set(custID_books)
```

```python
recommend(276812)
```
